# White Noise using the Monte Carlo Simulation

## Introduction
### In this Notebook we'll work on the White Noise process using the Monte Carlo simulation (Random Walk and Potential outcomes)
First, let's start by importing the neccessary libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

### Function to generate the Random Walk

```python
def generate_random_walk(n, mean=0, std_dev=1):
    return np.random.normal(mean, std_dev, n).cumsum()
```

### Monte Carlo Simulation Function

```python
def monte_carlo_simulation(num_simulations=1000, n=1000, mean=0, std_dev=1):
    simulated_series = []
    for _ in tqdm(range(num_simulations), desc="Monte Carlo Simulation"):
        simulated_series.append(generate_random_walk(n, mean, std_dev))
    return np.array(simulated_series)

def plot_monte_carlo_simulation(simulated_series):
    plt.figure(figsize=(12, 6))
    for simulation in simulated_series:
        plt.plot(simulation, linewidth=0.6, alpha=0.1)
    plt.title('Monte Carlo Simulation of Random Walk')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.grid(True)
    plt.show()
```

### Plot the Results

```python
# Main code
num_simulations = 1000
n = 1000

# Perform Monte Carlo simulation
simulated_series = monte_carlo_simulation(num_simulations=num_simulations, n=n)

# Plot the results
plot_monte_carlo_simulation(simulated_series)
```

### Understanding White Noise and Monte Carlo Simulation:
 **White Noise**: 
White noise is a sequence of random variables with a mean of zero and a constant variance, representing purely random fluctuations without any discernible pattern.

**Monte Carlo Simulation**: 
This is a statistical technique used to model and analyze the behavior of complex systems by simulating the outcomes of a process multiple times.

### Implementation Steps:
**Generating White Noise**: 
We started by generating white noise series to serve as the basis for our simulations.

**Simulation Process**: 
Using the generated white noise, we conducted multiple Monte Carlo simulations to create a set of possible future paths for our process. Each path represents a potential future outcome based on the random variations from white noise.

### Results Analysis:

### Statistical Insights:
**Mean**:

**Variance and Standard Deviation**: 
These measures quantify the dispersion and volatility of the simulated outcomes. Higher values indicate greater uncertainty and wider potential ranges for future outcomes.
